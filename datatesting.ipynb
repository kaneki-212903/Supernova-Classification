```python
#Calculating distance in light year using redshift

def distance(redshift):
    speed_of_light=299792458
    Hubble_constant=70
    megaparsec_to_light_year=3.262*10**6

    distance_mega_parsec= (speed_of_light*redshift)/Hubble_constant
    distance_light_year=distance_mega_parsec*megaparsec_to_light_year
    
    return distance_light_year
```

```python
#Calculating coordinates x,y,z using redshift and distance and make a new column
import math
def calculate_cartesian_coordinates(ra, decl, distance_light_years):
    
    # Convert angles from degrees to radians
    ra_rad = math.radians(ra)
    decl_rad = math.radians(decl)

    # Calculate Cartesian coordinates
    x = distance_light_years * math.cos(ra_rad) * math.cos(decl_rad)
    y = distance_light_years * math.sin(ra_rad) * math.cos(decl_rad)
    z = distance_light_years * math.sin(decl_rad)

    return x, y, z
```

```python
#Calculating coordinates using ra and decl and make a new column
data['x_coordinate']=distance(data['REDSHIFT_FINAL'])*np.cos(np.radians(data['RA']))*np.cos(np.radians(data['DECL']))
data['y_coordinate']=distance(data['REDSHIFT_FINAL'])*np.sin(np.radians(data['RA']))*np.cos(np.radians(data['DECL']))
data['z_coordinate']=distance(data['REDSHIFT_FINAL'])*np.sin(np.radians(data['DECL']))

data.head()
```

Let us also test using neural networks.

```python
# Create a classification neural network model to classify the supernovae into 7 different lcasses
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam

tf.random.set_seed(1008)

model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(7, activation='linear')
], name='SupernovaeClassifier')

model.compile(optimizer=Adam(learning_rate=0.001), loss=SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train_encoded, epochs=200, batch_size=64, validation_data=(X_cv, y_cv_encoded))

# Plot the training and validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Get the test accuracy
test_loss, test_accuracy = model.evaluate(X_test, y_test_encoded)
print(f"Test Accuracy = {test_accuracy}")
```

```python
# Get the precision, recall and f1-score
y_cv_pred = model.predict(X_cv)
y_cv_pred = np.argmax(y_cv_pred, axis=1)

classification_report_test = classification_report(y_cv_encoded, y_cv_pred, digits=5)
print(classification_report_test)
```
